```
from mpl_toolkits import mplot3d

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np

fig = plt.figure(100)

fig = plt.figure(figsize=(40,40))

u = np.linspace(0, np.pi, 100)
v = np.linspace(0, np.pi, 100)
y = 10 * np.outer(np.cos(u), np.sin(v))
x = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))

ax1 = fig.add_subplot(4, 4, 1, projection='3d')
surf = ax1.plot_surface(x, y, z, color='b')

ax2 = fig.add_subplot(4, 4, 2)
ax2.contour(x, y, z, colors='blue')

ax3 = fig.add_subplot(4, 4, 3)
ax3.contour(x, y, z, 30, cmap='RdGy')

ax1.scatter(0, 0, 10, c = 'r')
ax2.scatter(0,0, c = 'r')
ax3.scatter(0,0, c = 'r')


ax4 = fig.add_subplot(4, 4, 5, projection='3d')
surf = ax4.plot_surface(y, x, z, color='g')

ax5 = fig.add_subplot(4, 4, 6)
ax5.contour(y, x, z, colors='green')

ax6 = fig.add_subplot(4, 4, 7)
ax6.contour(y, x, z, 30, cmap='RdGy')

ax4.scatter(0, 0, -10, c = 'b')
ax5.scatter(0,0, c = 'b')
ax6.scatter(0,0, c = 'b')


ax7 = fig.add_subplot(4, 4, 9, projection='3d')
surf = ax7.plot_surface(x, y, y, color='r')

ax8 = fig.add_subplot(4, 4, 10)
ax8.contour(x, y, y, colors='red')

ax9 = fig.add_subplot(4, 4, 11)
ax9.contourf(x, y, y, 30, cmap='RdGy')

ax7.scatter(0, 0, 0, c = 'g')
ax8.scatter(0,0, c = 'g')
ax9.scatter(0,0, c = 'g')

print("Figures go from 1 -> 3 up -> down, as do the corresponding points on each")

plt.show()
```
